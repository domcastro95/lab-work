```python
# read in the file and make dictionaries of DBH at each cenus

in_file = open("/home/eeb177-student/Desktop/eeb-177/class-assignments/classwork-21-Feb/cocoli.txt")
sizes_94 = {}
sizes_97 = {}
sizes_98 = {}

# get rid of the header row and start on the next row 
in_file.readline()

# work 
for line in in_file:
    items = line.split()
    #retrieve relevent columns 
    tree_id = items[0] #now we are going into the data and pick which data to use
    s94 = float(items[4])
    s97 = float(items[5])
    s98 = float(items[6])
    
    # populate our dictionaries
    sizes_94[tree_id] = s94
    sizes_97[tree_id] = s97
    sizes_98[tree_id] = s98
in_file.close()
```

```python
# create a histo gram 
# go to matplotlib library and get the pyplot function.. name pyplot as plt

import matplotlib.pyplot as plt

#let's make a histogram of DBHs in 1994
# matplotlib works on lists, so let's make one
sizes_in_1994 = list(sizes_94.values())
```

```python
# 'hist = histogram' 
# show means show the plt'-d graph 
# when add 'log = True' changes the graph into 10^x, this graph is more useful 
plt.hist(sizes_in_1994, log = True)
plt.show()
```

```python
sizes_in_1994 = list(sizes_94.values()) #grab listed values 
sizes_in_1998 = list(sizes_98.values())

f, (ax1, ax2) = plt.subplots(1,2,sharex = True, sharey = True) # use the same 'x' axis for both plots 
ax1.hist(sizes_in_1994, log = True)
ax1.set_title('Hist_in_1994')
ax2.hist(sizes_in_1998, log = True)
ax2.set_title('Hist_in_1998')
plt.show()
```
